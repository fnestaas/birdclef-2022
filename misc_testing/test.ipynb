```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
data1 = .1*(np.linspace(0, 10, 10) + np.random.normal(0, 1, (10, )))
data2 = 1 - data1
```

```python
plt.plot(np.linspace(0, 10, 10), data1, label='F1 Macro')
plt.plot(np.linspace(0, 10, 10), data2, label='F1 Micro')
plt.legend()
plt.xlabel('Ablation Index')
plt.ylabel('Score')
plt.show()
```
